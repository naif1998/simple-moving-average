### Import necessary libraries

```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
```

### Ask user to input stock ticker

```python
stock= input("Enter Stock Ticker: ")
ma=requests.get("https://financialmodelingprep.com/api/v3/historical-price-full/"+stock+"?serietype=line&apikey=c543d9cd0eb889b3e1550a91e421eda6")
ma=ma.json()
ma=ma["historical"][0:900]

ma=pd.DataFrame.from_dict(ma)
ma=ma.set_index('date')
```

### Rolling is usually used in time-series data


```python
ma['20 days']=ma['close'].rolling(20,min_periods=1).mean()
ma['200 days']=ma['close'].rolling(200,min_periods=1).mean()
```

### Visualization

```python
ma[['close','20 days', '200 days']].plot(figsize=(10,10)).invert_xaxis()
plt.grid(True)
plt.ylabel('Price')
plt.title(stock + " Moving Averages")
```
